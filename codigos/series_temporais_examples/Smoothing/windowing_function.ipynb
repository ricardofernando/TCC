```python
### Windowing Function

import pandas as pd
import numpy as np

df = pd.read_csv('istambul_stock_exchange.txt', parse_dates=['date'])

df
```

```python
import numpy as np
import numpy_ext as npext


def coef_func(x,y):
    coef = np.polyfit(x,y,1)
    return coef[0]

df['fit'] = npext.rolling_apply(coef_func, 4, df.SP.values, df.DAX.values)

df
```

```python
#Rolling Window - MÃ©dia
df['SP_mean'] = df['SP'].rolling(window=4).mean()
```
